```python
import pandas as pd

# Load dataset
file_path = "./netflix_users.csv"  # Update the path if needed
df = pd.read_csv(file_path)

# Display first 5 rows
df.head()
```

```python
# Get summary statistics of numeric columns
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.countplot(
    data=df, 
    x="Subscription_Type", 
    hue="Subscription_Type",  # Explicitly set hue
    palette="pastel", 
    order=df["Subscription_Type"].value_counts().index, 
    legend=False  # Disable legend
)
plt.title("Subscription Type Distribution", fontsize=14)
plt.xlabel("Subscription Type", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=df, 
    x="Subscription_Type", 
    y="Watch_Time_Hours", 
    hue="Subscription_Type",  # Explicitly set hue
    palette="muted", 
    legend=False  # Disable legend
)
plt.title("Watch Time Distribution by Subscription Type", fontsize=14)
plt.xlabel("Subscription Type", fontsize=12)
plt.ylabel("Watch Time (Hours)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(
    data=df, 
    y="Favorite_Genre", 
    hue="Favorite_Genre",  # Explicitly set hue
    palette="coolwarm", 
    order=df["Favorite_Genre"].value_counts().index, 
    legend=False  # Disable legend
)
plt.title("Most Preferred Genres", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x="Age", y="Watch_Time_Hours", hue="Subscription_Type", palette="viridis", alpha=0.7)
plt.title("Age vs Watch Time", fontsize=14)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Watch Time (Hours)", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.countplot(
    data=df, 
    x="Country", 
    hue="Country",  # Explicitly set hue
    palette="Set2", 
    order=df["Country"].value_counts().index, 
    legend=False  # Disable legend
)
plt.title("User Distribution by Country", fontsize=14)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
