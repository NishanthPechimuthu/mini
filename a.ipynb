```python
import pandas as pd

# Load the dataset
df = pd.read_csv("netflix_data.csv")

# Display the first 5 rows
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot subscription type distribution
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Subscription_Type', hue='Subscription_Type', palette='Set2', legend=False)
plt.title("Subscription Type Distribution")
plt.xlabel("Subscription Type")
plt.ylabel("Count")
plt.show()
```

```python
# Average watch time per subscription type
df.groupby('Subscription_Type')['Watch_Time_Hours'].mean().plot(kind='bar', color=['blue', 'green', 'red'])
plt.title("Average Watch Time per Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Average Watch Time (Hours)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count occurrences of each genre
genre_counts = df['Favorite_Genre'].value_counts()

# Convert to DataFrame
genre_df = pd.DataFrame({'Genre': genre_counts.index, 'Count': genre_counts.values})

# Plot top 5 genres
plt.figure(figsize=(8, 4))
sns.barplot(data=genre_df.head(5), x='Genre', y='Count', hue='Genre', dodge=False, palette='viridis', legend=False)
plt.title("Top 5 Most Popular Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```
